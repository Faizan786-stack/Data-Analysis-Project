```python
import pandas as pd
```

```python
orange = pd.read_csv('orange_cap.csv')
purple = pd.read_csv('purple_cap.csv')
```

```python
orange
```

```python
purple.head()
```

```python
orange_counts = orange['Batsman'].value_counts()
```

```python
orange_counts
```

```python
purple_counts = purple['Bowler'].value_counts()
purple_counts
```

## Which player appeared most frequently in Orange or Purple Cap lists?

```python
combined = orange_counts.add(purple_counts, fill_value=0)
combined
```

```python
most_capped_players = combined.sort_values(ascending=False)
most_capped_players.head()
```

## Compare top batsmen vs top bowlers over seasons.

```python
top_batsmen = orange.groupby('Highest_score')['Batsman'].value_counts().sort_index(ascending=False).head(5)
top_batsmen
```

```python
top_bowler = purple.groupby('Bowler')['Wickets'].value_counts().sort_values(ascending=False).head(5)
```
