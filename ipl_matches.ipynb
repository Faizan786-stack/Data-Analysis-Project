## Which team has won the most matches overall?

```python
winner = matches["match_winner"].value_counts()
```

```python
winner
```

```python
date = matches['match_result'].count()
date
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(winner.index, winner.values)
plt.xlabel('Team')
plt.ylabel('Winner')
plt.title('WON MOST OF MATCHES')
plt.show()
```

## How many matches were played each season?

```python
matches_per_team = matches['team1'].value_counts() + matches['team2'].value_counts()
matches_per_team.sort_values()
```

```python
plt.scatter(matches_per_team.index, matches_per_team.values)
plt.title('Team Played Matches Each Seoson')
plt.xlabel('Team')
plt.ylabel('Matches')
plt.grid(True)
plt.show()
```

## Which venue hosted the most matches?

```python
venue = matches['venue'].str.strip().value_counts()
venue
```

```python
plt.barh(venue.index, venue.values, color='skyblue')
plt.title('venue')
plt.xlabel('Number of Matches')
plt.ylabel('Match Venue')
plt.gca().invert_yaxis()
plt.show()
```

## Who won the most "Player of the Match" awards?

```python
POF = matches['player_of_the_match'].value_counts().head(5)
```

```python
POF
```

```python
plt.figure(figsize=(8,5))
plt.bar(POF.index, POF.values, color='#8aec55')
plt.title('Most PLayer Of The Match')
plt.ylabel('Awards')
plt.xlabel("Player")
plt.xticks(rotation=45)
plt.show()
```
