## Which batsman scored the most runs?

```python
most_run = most_run.sort_values(ascending=False)
```

```python
MR
```

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(MR.index, MR.values, color='#54f3e0')
plt.title('MOST RUN OF THE BATSMAN')
plt.xlabel('Runs')
plt.ylabel('Player Name')
plt.gca().invert_yaxis()
plt.show()
```

## Which bowler has the most wickets?

```python
df2.fillna(0).head()
```

```python
wikets = df2.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False)
wikets
```

```python
top_wikets = wikets.drop_duplicates().head(10)
top_wikets
```

```python
plt.figure(figsize=(10,5))
plt.bar(top_wikets.index, top_wikets.values, color='#55f731')
plt.title('Top 10 Most Of The Wickets')
plt.xlabel('Bowlers')
plt.ylabel('Wickets')
plt.xticks(rotation=45)
plt.show()
```

## Who took the most catches/runouts/stumpings?

```python
wicket_type = df2['wicket_type'].value_counts()
```

```python
wicket_type
```

```python
plt.figure(figsize=(10,6))
plt.bar(wicket_type.index, wicket_type.values, color='#f7af31')
for i in range(len(wicket_type)):
    plt.text(i, wicket_type.values[i] + 10, str(wicket_type.values[i]), ha='center')
plt.title('most catches/runouts/stumpings')
plt.xlabel('Wickets Type')
plt.ylabel('Number Of Wickets')
plt.xticks(rotation=45)
plt.show()
```

## Which bowlers have the most dot balls in the powerplay (overs 1â€“6)?

```python
power_play = df2[df2['over'].between(1,6)]
dot_balls = power_play[power_play['runs_of_bat'] == 0]
dot_balls_counts = dot_balls.groupby('bowler').size().reset_index(name='dot_balls')
top_dots =  dot_balls_counts.sort_values(by='dot_balls', ascending=False)
```

```python
most_dots = top_dots.head(10).reset_index().drop('index', axis=1)
```

```python
most_dots
```

```python
plt.figure(figsize=(10, 6))
plt.barh(most_dots['bowler'], most_dots['dot_balls'], color='#74f8fa')
plt.title('Top 10 Bowlers with Most Dot Balls in Powerplay (Overs 1â€“6)', fontsize=14)
plt.xlabel('Dot Balls')
plt.ylabel('Bowler')
for index, value in enumerate(most_dots['dot_balls']):
    plt.text(value + 1, index, str(value), va='center')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Which bowler has dismissed Virat Kohli the most?

```python
v_k = df2.query("striker == 'Kohli'").copy()
```

```python
v_k.head()
```

```python
dismiss_v_k = v_k[['striker','bowler']].value_counts().dropna().reset_index().head(6)
dismiss_v_k
```

```python
dismiss_v_k.columns = ['striker','bowler', 'value']
```

```python
labels = dismiss_v_k['bowler']
values = dismiss_v_k['value']
explode = [0.1,0,0,0,0,0]
colors = ['#40ebf9','blue','#40f982','#fc0d0d','#de0dfc','#fcf25d']
plt.pie(values, labels=labels, explode=explode, autopct='%1.1f%%', colors=colors,shadow=True)
plt.title('Bowler Has Dismissed Virat Kohli')
plt.tight_layout()
plt.show()
```
